```python
d1 = {
    "brand": "ford",
    "model":"mustang",
    "year": 1964,
    "year": 2020
}
print(d1)


#changing a value in the dictionary
d2 = {
    "brand": "ford",
    "model":"mustang",
    "year": 1964,
}
d2["year"] = 2024
print(d2)

#adding a new key value
d2["color"] = "red"
print(d2)

#changing a value using .update() and printing the value using .values()
d2.update({"color": "pink"})
y = d2.values()
print(y)

#length of the dictionary
dd = len(d2)
print ("length of the dictionary: ", dd)
```

```python
# program 2:
# a dictionary stores student names and marks
# write a python program to search for a student's marks using the name entered by the user
# if the student is not found, display a suitable message
d = {"samarth": 10, "rudransh": 9, "riya": 8 }
y = input("enter the name of the student: ")
if y in d:
  print ("the marks of the student are: ", d[y])
else:
  print(" the student isnt in the dictionary")
```
